```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from paretoset import paretoset
from pathlib import Path
```

## 1. Load the data.

The data is located in *outputs/dataset_Coursework_Question4.csv.


```python
savePath = Path("supplementaryData", "dataset_Coursework_Question4.csv")
df = pd.read_csv(savePath, index_col=0)
print(df.head())
```

### 1.1 Convert the raw data into more practical units
The data should contain four columns with building attributes (U-value of windows, g-value of windows, WWR, and the fixed shading depth) plus four columns of performance data. These are the annual sums of lighting and heating energy usage, the peak heating load, and the number of hours above 28C inside the building.

<span style="color:orange;"> Remember that EnergyPlus outputs are in base SI units. </span>

* lightingSum is given in Joules. Convert --> MJ
* heatingSum is given in Joules. Convert --> MJ
* heatingMax is given in Watts. Convert --> kW
* temperature>28C is given in hours.

NOTE: That we won't be using the output *heatingMax* for this exercise.

```python
# Convert EnergyPlus outputs from Joules → MJ, Watts → kW
df["heatingSum"] /= 10**6
df["lightingSum"] /= 10**6
df["heatingMax"] /= 10**3

print("The first 10 rows of the dataframe look like this:")
print(df.head(10))
```

## 2. Analysis

<span style="color:limegreen;"> Complete the following analysis sections. You can re-use the code in exercise 4A and 4B to help complete the analysis.</span>

### 2.1 Pareto Front Analysis

<span style="color:limegreen;"> Perform the pareto front analysis for the three bi-objective problems and the tri-objective problems here. You are free to improve upon the analysis and visualisations too.</span>

#### 2.1.1 Heating and Lighting

#### 2.1.1 Heating and Lighting

```python
columns = ["heatingSum", "lightingSum"]
pareto_front = paretoset(df[columns], sense=["min", "min"])

onPareto = df[pareto_front]
offPareto = df[~pareto_front]

fig, ax = plt.subplots(ncols=2, figsize=(12,5))
ax[0].scatter(offPareto["heatingSum"], offPareto["lightingSum"], c="red", edgecolors="black", alpha=0.6)
ax[0].scatter(onPareto["heatingSum"], onPareto["lightingSum"], c="lime", edgecolors="black", label="Pareto Front")
ax[0].set_xlabel("Heating Energy (MJ)")
ax[0].set_ylabel("Lighting Energy (MJ)")
ax[0].set_title("Pareto Front – Heating vs Lighting")
ax[0].legend()

ax[1].bar(["Total Points", "On Pareto Front"], [len(df), pareto_front.sum()], color=["gray", "lime"])
ax[1].set_title("Pareto Front Summary")
plt.tight_layout()
plt.show()
```

#### 2.1.2 Heating and Overheating

```python
columns = ["heatingSum", "temperature>28C"]
pareto_front = paretoset(df[columns], sense=["min", "min"])

onPareto = df[pareto_front]
offPareto = df[~pareto_front]

fig, ax = plt.subplots(figsize=(7,5))
ax.scatter(offPareto["heatingSum"], offPareto["temperature>28C"], c="red", edgecolors="black", alpha=0.6)
ax.scatter(onPareto["heatingSum"], onPareto["temperature>28C"], c="lime", edgecolors="black", label="Pareto Front")
ax.set_xlabel("Heating Energy (MJ)")
ax.set_ylabel("Overheating Hours > 28°C")
ax.set_title("Pareto Front – Heating vs Overheating")
ax.legend()
plt.tight_layout()
plt.show()
```

#### 2.1.3 Overheating and Lighting

```python
columns = ["temperature>28C", "lightingSum"]
pareto_front = paretoset(df[columns], sense=["min", "min"])

onPareto = df[pareto_front]
offPareto = df[~pareto_front]

fig, ax = plt.subplots(figsize=(7,5))
ax.scatter(offPareto["temperature>28C"], offPareto["lightingSum"], c="red", edgecolors="black", alpha=0.6)
ax.scatter(onPareto["temperature>28C"], onPareto["lightingSum"], c="lime", edgecolors="black", label="Pareto Front")
ax.set_xlabel("Overheating Hours > 28°C")
ax.set_ylabel("Lighting Energy (MJ)")
ax.set_title("Pareto Front – Overheating vs Lighting")
ax.legend()
plt.tight_layout()
plt.show()
```

#### 2.1.4 Heating, Lighting, and Overheating

```python
columns = ["heatingSum", "lightingSum", "temperature>28C"]
pareto_front = paretoset(df[columns], sense=["min", "min", "min"])

onPareto = df[pareto_front]
print(f"{pareto_front.sum()} points are on the Pareto front out of {len(df)} total points.")
print(onPareto[columns].head())

# Larger figure size and clearer layout
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection="3d")

# Plot all points (non-dominated)
ax.scatter(df["heatingSum"], df["lightingSum"], df["temperature>28C"],
           c="red", alpha=0.5, s=30, label="Non-dominated")

# Highlight Pareto Front
ax.scatter(onPareto["heatingSum"], onPareto["lightingSum"], onPareto["temperature>28C"],
           c="lime", s=80, edgecolors="black", label="Pareto Front")

# Axis labels and title
ax.set_xlabel("Heating (MJ)", labelpad=12)
ax.set_ylabel("Lighting (MJ)", labelpad=12)
ax.set_zlabel("Overheating (>28°C Hours)", labelpad=12)
ax.set_title("3D Pareto Front – Heating, Lighting, Overheating", pad=20)

# Adjust viewpoint and tick size
ax.view_init(elev=25, azim=135)
ax.tick_params(axis='both', which='major', labelsize=10)
ax.legend(loc="upper right", fontsize=10)

# Add extra padding
plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05)
plt.show()
```

### 2.2 Weighted Sum Analysis

#### 2.2.1 Heating and Lighting

```python
w = [0.5, 0.5]  # equal weighting
df["WeightedSum"] = w[0]*df["heatingSum"] + w[1]*df["lightingSum"]

min_idx = df["WeightedSum"].idxmin()
best_point = df.loc[min_idx]

print("Optimal (equal weights):")
print(best_point[["heatingSum", "lightingSum"]])

fig, ax = plt.subplots(figsize=(7,5))
ax.scatter(df["heatingSum"], df["lightingSum"], c="red", alpha=0.6)
ax.scatter(best_point["heatingSum"], best_point["lightingSum"], c="deepskyblue", s=80, label="Weighted Sum Optimum")
ax.set_xlabel("Heating (MJ)")
ax.set_ylabel("Lighting (MJ)")
ax.legend()
ax.set_title("Weighted Sum – Heating vs Lighting")
plt.show()
```

#### 2.2.2 Heating and Overheating

```python
w = [0.5, 0.5]
df["WeightedSum"] = w[0]*df["heatingSum"] + w[1]*df["temperature>28C"]

min_idx = df["WeightedSum"].idxmin()
best_point = df.loc[min_idx]

print("Optimal (equal weights):")
print(best_point[["heatingSum", "temperature>28C"]])

fig, ax = plt.subplots(figsize=(7,5))
ax.scatter(df["heatingSum"], df["temperature>28C"], c="red", alpha=0.6)
ax.scatter(best_point["heatingSum"], best_point["temperature>28C"], c="deepskyblue", s=80, label="Weighted Sum Optimum")
ax.set_xlabel("Heating (MJ)")
ax.set_ylabel("Overheating Hours > 28°C")
ax.legend()
ax.set_title("Weighted Sum – Heating vs Overheating")
plt.show()
```

#### 2.2.3 Overheating and Lighting

```python
w = [0.5, 0.5]
df["WeightedSum"] = w[0]*df["temperature>28C"] + w[1]*df["lightingSum"]

min_idx = df["WeightedSum"].idxmin()
best_point = df.loc[min_idx]

print("Optimal (equal weights):")
print(best_point[["temperature>28C", "lightingSum"]])

fig, ax = plt.subplots(figsize=(7,5))
ax.scatter(df["temperature>28C"], df["lightingSum"], c="red", alpha=0.6)
ax.scatter(best_point["temperature>28C"], best_point["lightingSum"], c="deepskyblue", s=80, label="Weighted Sum Optimum")
ax.set_xlabel("Overheating Hours > 28°C")
ax.set_ylabel("Lighting (MJ)")
ax.legend()
ax.set_title("Weighted Sum – Overheating vs Lighting")
plt.show()
```

#### 2.2.4 Heating, Lighting, and Overheating

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 (needed for 3D)

# --- 1) Select and normalize objectives (minimize all three) ---
obj_cols = ["heatingSum", "lightingSum", "temperature>28C"]

# Ensure numeric
df[obj_cols] = df[obj_cols].apply(pd.to_numeric, errors="coerce")
df = df.dropna(subset=obj_cols).copy()

# Min–max normalization to [0,1] (protect against zero range)
for c in obj_cols:
    rng = df[c].max() - df[c].min()
    if rng == 0:
        df[c + "_norm"] = 0.0
    else:
        df[c + "_norm"] = (df[c] - df[c].min()) / rng

# --- 2) Weighted sum (equal weights, normalized objectives) ---
w = np.array([1/3, 1/3, 1/3], dtype=float)
w = w / w.sum()  # just in case

df["WeightedSum"] = (
    w[0] * df["heatingSum_norm"]
  + w[1] * df["lightingSum_norm"]
  + w[2] * df["temperature>28C_norm"]
)

# Best index (minimization). Handle ties gently.
min_val = df["WeightedSum"].min()
tol = 1e-12
best_idxs = df.index[np.isclose(df["WeightedSum"], min_val, atol=tol)]
best_idx = best_idxs[0]
best_point = df.loc[best_idx]

print("Optimal (equal weights, normalized tri-objective):")
print(best_point[["heatingSum", "lightingSum", "temperature>28C"]])
print("Weighted sum (normalized):", float(min_val))

# --- 3) Plot ---
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection="3d")

# All points
ax.scatter(
    df["heatingSum"], df["lightingSum"], df["temperature>28C"],
    c="red", alpha=0.5, s=30, label="All Points", zorder=1
)

# Highlight ALL ties (if any)
ax.scatter(
    df.loc[best_idxs, "heatingSum"],
    df.loc[best_idxs, "lightingSum"],
    df.loc[best_idxs, "temperature>28C"],
    s=220, facecolors="none", edgecolors="blue",
    linewidths=2.5, zorder=10, label="Weighted Sum Optimum"
)

# Axis labels and formatting
ax.set_xlabel("Heating (MJ)", labelpad=12)
ax.set_ylabel("Lighting (MJ)", labelpad=12)
ax.set_zlabel("Overheating (>28°C hours)", labelpad=12)
ax.set_title("Weighted Sum – 3D Tri-objective", pad=20)

# View angle helps separation
ax.view_init(elev=25, azim=135)
ax.tick_params(axis="both", which="major", labelsize=10)
ax.legend(loc="upper right", fontsize=10)

plt.subplots_adjust(left=0.05, right=0.95, top=0.95, bottom=0.05)
plt.show()
```
